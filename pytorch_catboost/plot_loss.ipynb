```python
import numpy as np
import matplotlib.pyplot as plt

from torch import Tensor

from pytorch_catboost.playground_regression import _custom_loss
import pandas as pd
import pandas_bokeh
pandas_bokeh.output_notebook()


raw_diff = np.linspace(-1, 1, 101)
loss = _custom_loss(Tensor(raw_diff)).detach().numpy()
```

```python
series = pd.Series(data=loss, index=raw_diff
                  ).plot_bokeh(legend=None,
                               xticks=np.arange(-100,100.01,20)/100,
                               xlabel="diff = (preds - targets)",
                               ylabel="loss",
                               title="Custom regression loss function - don't undershoot!",
                               fontsize_label=14,
                               fontsize_title=14,
                               fontsize_ticks=11)
```
