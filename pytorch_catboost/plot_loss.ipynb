```python
import numpy as np
import matplotlib.pyplot as plt

from torch import Tensor

from pytorch_catboost.playground_regression_for_medium import custom_regression_loss, very_custom_regression_loss
import pandas as pd
import pandas_bokeh
pandas_bokeh.output_notebook()
```

```python
def plot_me(loss, preds):
    return pd.Series(data=loss, index=preds
                  ).plot_bokeh(legend=None,
                               xticks=np.arange(-100,100.01,20)/100,
                               xlabel="diff = (preds - targets)",
                               ylabel="loss",
                               title="Custom regression loss function - don't undershoot!",
                               fontsize_label=14,
                               fontsize_title=14,
                               fontsize_ticks=11)
```

```python
preds = np.linspace(-1, 1, 101)
targets = np.zeros_like(preds)

loss = np.hstack([custom_regression_loss(Tensor([p]), Tensor([t])).detach().numpy()
                  for p, t in zip(preds, targets)])

another_loss = np.hstack([very_custom_regression_loss(Tensor([p]), Tensor([t])).detach().numpy()
                  for p, t in zip(preds, targets)])
```

```python
display(plot_me(loss, preds))
display(plot_me(another_loss, preds))
```
